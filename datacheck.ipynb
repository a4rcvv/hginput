```python
import polars as pl
import seaborn as sns
```

```python
def show_df_stats(path: str):
    df = pl.read_parquet(path)
    print(df.shape)
    sns.countplot(x="label", data=df.to_pandas())
```

```python
show_df_stats("./hginput/model/data/merged/merged_train.parquet.zstd")
```

```python
show_df_stats("./hginput/model/data/merged/merged_test.parquet.zstd")
```

```python
show_df_stats("./hginput/model/data/v1/v1_test.parquet.zstd")
```
